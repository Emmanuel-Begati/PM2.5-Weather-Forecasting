# Separate features and target

- Feel free to trop any non-essential columns like that you think might not contribute to modeling.

```python
X_train = train.drop(['pm2.5', 'No'], axis=1)
y_train = train['pm2.5']
```

```python
# Reshape data for LSTM input
# LSTM models require data in the shape (samples, timesteps, features).
# Here, the data is reshaped to add a "timesteps" dimension.
X_train = np.expand_dims(X_train, axis=1)
```

# Build model

Below is a simple LSTM model. Your task is to experiment with different parameters like, numbers of layers, units, activation functions, and optimizers, etc to get the best performing model. Experiment with other optimizers (e.g., SGD) or hyperparameters to improve performance.

```python
# Define model
model = Sequential([
    # First LSTM layer
    LSTM(128, activation='tanh', return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    BatchNormalization(),
    Dropout(0.3),

    # Second LSTM layer
    LSTM(64, activation='tanh', return_sequences=True),
    BatchNormalization(),
    Dropout(0.3),

    # Third LSTM layer
    LSTM(32, activation='tanh', return_sequences=False),
    BatchNormalization(),
    Dropout(0.2),

    # Fully connected layers
    Dense(64, activation='relu'),
    Dropout(0.2),

    Dense(32, activation='relu'),

    # Output layer
    Dense(1)  # Output for regression
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.0003),
    loss='mse',
    metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')]
)

# Display model architecture
model.summary()

# Callbacks for early stopping and learning rate reduction
callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1),
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=1)
]
```

```python
# Train the model
history = model.fit(
    X_train, y_train,
    epochs=70,
    batch_size=32,
    validation_split=0.2,
    callbacks=callbacks
)

# Training RMSE
train_rmse = history.history['rmse'][-1]
print(f"Training RMSE: {train_rmse}")
```

```python
print(test.columns)
```

```python
# Calculate training loss
train_predictions = model.predict(X_train)
train_loss = np.mean((y_train - train_predictions.flatten())**2)

print(f"Test Loss (MSE): {train_loss}")

# Plot training loss
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')  # Training loss during epochs
plt.axhline(y=train_loss, color='blue', linestyle='--', label='Final rain Loss')  # Final training loss
plt.title('Loss on Training Data')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()

print(f"Final Training Loss (MSE): {train_loss}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Make predictions on training data
train_predictions = model.predict(X_train)

# Calculate MSE (Mean Squared Error)
train_mse = mean_squared_error(y_train, train_predictions)
print(f"Training MSE: {train_mse}")

# Calculate MAE (Mean Absolute Error)
train_mae = mean_absolute_error(y_train, train_predictions)
print(f"Training MAE: {train_mae}")

# Calculate R² score
train_r2 = r2_score(y_train, train_predictions)
print(f"Training R²: {train_r2}")
```

```python
# Load sample submission file
sample_submission = pd.read_csv('sample_submission.csv')  # Removed extra space in filename

# Prepare the test data
X_test = test.drop(columns=['No'])

# Ensure correct shape: (samples, timesteps, features)
time_steps = X_train.shape[1]  # Extract from training set
num_features = X_train.shape[2]

X_test = X_test.values.reshape((-1, time_steps, num_features))

print(f"Reshaped X_test shape: {X_test.shape}")

# Make predictions
predictions = model.predict(X_test)

# Ensure no NaN values
predictions = np.nan_to_num(predictions)

# Convert predictions to integers
predictions = np.round(predictions).astype(int)

# Prepare submission DataFrame
submission = pd.DataFrame({
    'row ID': test.index.astype(str),  # Ensuring correct formatting
    'pm2.5': predictions.flatten()
})

# Align with sample submission
submission = sample_submission[['row ID']].merge(submission, on='row ID', how='left')

# Fill missing values with 0
submission['pm2.5'].fillna(0, inplace=True)

# Save submission file
submission.to_csv('subm_fixed_4.csv', index=False)

print("Submission file saved successfully!")
```
